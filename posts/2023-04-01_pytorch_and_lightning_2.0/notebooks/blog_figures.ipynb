```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
exp1, exp2 = pd.read_csv('/content/exp1_results.csv'), pd.read_csv('/content/exp2_results.csv')
exp3, exp4 = pd.read_csv('/content/lightning.csv'), pd.read_csv('/content/lightning_compile.csv')
exp5, exp6 = pd.read_csv('/content/exp5_results.csv'), pd.read_csv('/content/exp6_results.csv')
exp7, exp8 = pd.read_csv('/content/exp7_results.csv'), pd.read_csv('/content/exp8_results.csv')
```

```python
epochs = np.array(range(len(exp1)))
label_fs = 14
title_fs = 16
stitle_fs = 18
```

```python
#| label: fig-pytorch-exp
#| fig-cap: Benchmark Comparison between PyTorch Eager and Compile Mode for 3 Epochs.
plt.figure(figsize=(25, 8))
plt.subplot(1, 3, 1)
plt.plot(epochs, exp1['train_loss'], label='eager_train_loss')
plt.plot(epochs, exp1['test_loss'], label='eager_val_loss')
plt.plot(epochs, exp2['train_loss'], label='compile_train_loss')
plt.plot(epochs, exp2['test_loss'], label='compile_val_loss')
plt.title('Loss', fontsize=title_fs)
plt.xlabel('Epochs', fontsize=label_fs)
plt.legend();
plt.subplot(1, 3, 2)
plt.plot(epochs, exp1['train_acc'], label='eager_train_acc')
plt.plot(epochs, exp1['test_acc'], label='eager_val_acc')
plt.plot(epochs, exp2['train_acc'], label='compile_train_acc')
plt.plot(epochs, exp2['test_acc'], label='compile_val_acc')
plt.title('Accuracy', fontsize=title_fs)
plt.xlabel('Epochs', fontsize=label_fs)
plt.legend();
plt.subplot(1, 3, 3)
plt.plot(epochs, exp1['train_epoch_time(min)'], label='eager_train_time')
plt.plot(epochs, exp1['test_epoch_time(min)'], label='eager_val_time')
plt.plot(epochs, exp2['train_epoch_time(min)'], label='compile_train_time')
plt.plot(epochs, exp2['test_epoch_time(min)'], label='compile_val_time')
plt.title('Time (min)', fontsize=title_fs)
plt.xlabel('Epochs', fontsize=label_fs)
plt.legend()
plt.suptitle('PyTorch Eager and Compile Mode', fontsize=stitle_fs, fontweight='bold');
```

```python
exp3_dict = []
for i, dfg in exp3.groupby('epoch'):
    agg = dict(dfg.mean())
    agg['epoch'] = i
    exp3_dict.append(agg)
exp4_dict = []
for i, dfg in exp3.groupby('epoch'):
    agg = dict(dfg.mean())
    agg['epoch'] = i
    exp4_dict.append(agg)
exp3, exp4 = pd.DataFrame(exp3_dict), pd.DataFrame(exp4_dict)
```

```python
#| label: fig-pytorch-light-exp
#| fig-cap: Benchmark Comparison between PyTorch Lightning Eager and Compile Mode for 3 Epochs.
plt.figure(figsize=(20, 6))
plt.subplot(1, 2, 1)
plt.plot(epochs, exp3['train_loss'], label='eager_train_loss')
plt.plot(epochs, exp3['test_loss'], label='eager_val_loss')
plt.plot(epochs, exp4['train_loss'], label='compile_train_loss')
plt.plot(epochs, exp4['test_loss'], label='compile_val_loss')
plt.title('Loss', fontsize=label_fs)
plt.xlabel('Epochs', fontsize=label_fs)
plt.legend();
plt.subplot(1, 2, 2)
plt.plot(epochs, exp3['train_acc'], label='eager_train_acc')
plt.plot(epochs, exp3['test_acc'], label='eager_val_acc')
plt.plot(epochs, exp4['train_acc'], label='compile_train_acc')
plt.plot(epochs, exp4['test_acc'], label='compile_val_acc')
plt.title('Accuracy', fontsize=label_fs)
plt.xlabel('Epochs', fontsize=label_fs)
plt.legend();
plt.suptitle('PyTorch Lightning Eager and Compile Mode', fontsize=title_fs, fontweight='bold');
```
